```python
from OBsim import *
plotly.offline.init_notebook_mode(connected=True)
```

```python
# Instantiate simulation with specific parameters
sim = Simulation(starting_infected=1,infected_growth_rate=0.1, awareness=0.001,
                 awareness_growth_rate=0.05, mortality_day_rate=0.01,
                 recovery_day_rate=0.01,recovery_multiplier=3,
                 containment_effectiveness=0.95,cycles=200)
```

```python
# Run Simulation
sim.go()
```

```python
# Get data from simulation, can be exported as CSV with .to_csv('file.csv',index=False)
data = sim.export()
data.tail(3)
```

```python
# Plot line graph
sim.plot()
```

```python
# Interactive barplot/animation
sim.bar()
```
